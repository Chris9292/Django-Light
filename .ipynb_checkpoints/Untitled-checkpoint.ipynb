```python
import networkx as nx
from matplotlib import pyplot as plt

G = nx.grid_2d_graph(10,10)

plt.figure(figsize=(10,10))
pos = {(x,y):(x,y) for x,y in G.nodes()}
nx.draw(G, pos=pos, 
        node_color='limegreen', 
        with_labels=True,
        node_size=1000)
```

```python
astar_path = nx.shortest_paths.astar_path(G, source=(0,0), target=(9,0))
```

```python
def compute_shortest_path(G, source, target):
    return nx.shortest_paths.astar_path(G, source, target)

def visualize_path(G, path):
    plt.figure(figsize=(10,10))
    pos = {(x,y):(x,y) for x,y in G.nodes()}
    path_edges = list(zip(path,path[1:]))
    nx.draw(G, pos=pos, 
            node_color='limegreen', 
            with_labels=True,
            node_size=1000)
    nx.draw_networkx_nodes(G,pos,nodelist=path,node_color='r')
    #nx.draw_networkx_edges(G,pos,edgelist=path,edge_color='r',width=10)
```

```python
path = compute_shortest_path(G, source=(6,4), target=(9,7))
path

visualize_path(G, path)
```
